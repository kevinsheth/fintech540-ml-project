```python
fs = s3fs.S3FileSystem(anon=True)  # Set to `anon=False` if you have credentials
files = fs.ls('s3://fintech540-ml/equity_data/prices')
```

```python
files
```

```python
master_df = pd.DataFrame()

# Loop through each file and read it into a DataFrame
for i, file in enumerate(files):
    df = pd.read_csv(f's3://{file}')
    file_name = file.split('/')[-1]
    df_melted = pd.melt(df, id_vars=['Unnamed: 0'], var_name='Date', value_name='Price')
    df_melted['Timestamp'] = pd.to_datetime(df_melted['Date'] + ' ' + df_melted['Unnamed: 0'])
    df_melted.set_index('Timestamp', inplace=True)
    df_melted.drop(['Unnamed: 0', 'Date'], axis=1, inplace=True)

    df_melted.to_parquet(f"staging/prices/MELTED_{file_name.replace('csv', 'parquet.gzip')}", compression='gzip')
```
